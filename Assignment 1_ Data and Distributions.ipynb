```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import powerlaw
from math import sqrt
from math import log
from math import log10
import math
from scipy.stats import norm
```

```
mu = 2
sigma = 5
normal = np.random.normal(mu, sigma, 10000)
plt.hist(normal, bins=100)
plt.title("HISTOGRAM FOR NORMAL DISTRIBUTION")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
a1 = 2 - (5 * sqrt(3))
b1 = 2 + (5 * sqrt(3))
# print(a1, b1)
u1 = np.random.uniform(a1, b1, 10000)
plt.hist(u1, bins=100)
plt.title("HISTOGRAM FOR U1")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
url = 'https://raw.githubusercontent.com/sarahyurick/datasets/master/airport_routes.csv'
airport_routes = pd.read_csv(url)
airport_routes.head()
```

```
routes = airport_routes['NumberOfRoutes']
plt.hist(routes, bins=50)
plt.title("HISTOGRAM FOR ROUTES DISTRIBUTION")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
```

```
url = 'https://raw.githubusercontent.com/sarahyurick/datasets/master/movie_votes.csv'
movie_votes = pd.read_csv(url)
movie_votes.head()
```

```
votes = movie_votes['AverageVote']
plt.hist(votes, bins=10)
plt.title("HISTOGRAM FOR VOTES DISTRIBUTION")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
```

```
a_sum = 0
a_n = 3409
for i in routes:
  a_sum += log(i)
inverse_asum = 1/a_sum
a_alpha = 1 + (a_n*inverse_asum)

print("alpha: ", a_alpha)

a_power = np.random.power(a_alpha, size=a_n)
plt.hist(a_power, bins=100)
plt.title("HISTOGRAM FOR POWER LAW DISTRIBUTION OF ROUTES")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
```

```
a_mean = np.mean(routes)
a_lamb = 1/a_mean

print("lambda: ", a_lamb)

a_exp = np.random.exponential(a_lamb, size=a_n)
plt.hist(a_exp, bins=100)
plt.title("HISTOGRAM FOR EXPONENTIAL DISTRIBUTION OF ROUTES")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
```

```
a_a = min(routes)
a_b = max(routes)

print("a :", a_a, " b: ", a_b)

a_uniform = np.random.uniform(a_a, a_b, a_n)
plt.hist(a_uniform, bins=100)
plt.title("HISTOGRAM FOR UNIFORM DISTRIBUTION OF ROUTES")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
a_sigma = np.std(routes)

print("mu: ", a_mean, " sigma: ", a_sigma)

a_normal = np.random.normal(a_mean, a_sigma, a_n)
plt.hist(a_normal, bins=100)
plt.title("HISTOGRAM FOR NORMAL DISTRIBUTION OF ROUTES")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
url = 'https://raw.githubusercontent.com/sarahyurick/datasets/master/appstore_games.csv'
appstore_games = pd.read_csv(url)
appstore_games.head()
```

```
num_ratings = appstore_games['User Rating Count']
num_ratings = num_ratings[np.logical_not(np.isnan(num_ratings))]
num_ratings = num_ratings.astype(int)
print(min(num_ratings))
print(max(num_ratings))
```

```
last_digits = []
for i in num_ratings:
  last_digits.append(i%10)
  
plt.hist(last_digits, bins=10)
plt.title("HISTOGRAM FOR LAST DIGITS")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
first_digits = []
for i in num_ratings:
  number = i
  count = int(log10(number))
  first_digit = number // pow(10, count)
  first_digit = int(first_digit)
  first_digits.append(first_digit)
  
plt.hist(first_digits, bins=10)
plt.title("HISTOGRAM FOR FIRST DIGITS")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
plt.show()
```

```
f_sum = 0
f_n = len(num_ratings)
for i in first_digits:
  f_sum += log(i/5)
inverse_fsum = 1/f_sum
f_alpha = 1 + (f_n*inverse_fsum)

print("alpha: ", f_alpha)

dist = st.expon
args = dist.fit(first_digits)
test1 = st.kstest(first_digits, dist.cdf, args)
print(test1)

f_power = np.random.power(f_alpha, size=f_n)
plt.hist(f_power, bins=100)
plt.title("HISTOGRAM FOR POWER LAW DISTRIBUTION OF RATINGS")
plt.xlabel("Values")
plt.ylabel("Number of occurrences")
```
